```python
def pearson_permute(inputDF, 
                    sampleList_ref=['COHP_44959','COHP_44978','COHP_46947'], 
                    sampleList_topermute=['COHP_44940','COHP_44997','COHP_46927','COHP_46939','COHP_49205','COHP_49224'], 
                    gene_name='ENSMUSG00000000001', 
                    nbootstrap=10000
                   ):
    '''
    controls week0 'COHP_44959' 'COHP_44978' 'COHP_46947'
    samples week0 'COHP_44940' 'COHP_44997' 'COHP_46927' 'COHP_46939' 'COHP_49205' 'COHP_49224'
    '''
    id_r = np.flatnonzero(inputDF.index==gene_name)
    id_c = np.flatnonzero(inputDF.columns.isin(sampleList_ref))
    id_c_permute = np.flatnonzero(inputDF.columns.isin(sampleList_topermute))
    a = np.sort(inputDF.values[id_r, id_c])

    holder = np.zeros(nbootstrap)
    # always need to whiten since we are resampling
    #if whitened: # dont need to norm dotproduct, as vector norm=1
    #    cor_own = np.dot(a,a)
    #    for x in range(nbootstrap):
    #        b = np.sort(inputDF.values[id_r, np.random.choice(id_c_permute, len(sampleList_ref))])
    #        cor = np.dot(a,b) 
    #        holder[x] = cor-cor_own
    
    for x in range(nbootstrap):
        b = np.sort(inputDF.values[id_r, np.random.choice(id_c_permute, len(sampleList_ref))])
        cor = np.dot(a,b) / (np.linalg.norm(a)*np.linalg.norm(b))
        chance1 = np.sort( inputDF.loc[gene_name].sample(len(sampleList_ref)) )
        chance2 = np.sort( inputDF.loc[gene_name].sample(len(sampleList_ref)) )
        cor_chance = np.dot(chance1,chance2) / (np.linalg.norm(chance1)*np.linalg.norm(chance2)) # correlation by chance by drawing from empirical dist of the gene
        holder[x] = cor-cor_chance
    
    return holder
                          
                    
```
